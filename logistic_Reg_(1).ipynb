```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("insurance_data.csv")
```

```python
df.head(3)
```

```python
df.shape
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="age",y="bought_insurance",data=df)
plt.show()
```

```python
x=df[['age']]
y=df['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[25]])# as we perdict value give to data is showing no at age of 25 he/she purchase the
#insurance yes or no it shown by graph
# red line show yes or no
# using the single input.
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="age",y="bought_insurance",data=df)
sns.lineplot(x="age",y=lr.predict(x),data=df,color='red')
plt.show()
```
