```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv('salary_3.csv')
df.head()
```

```python
sns.kdeplot(df['Age'])
plt.show()
```

```python
sns.scatterplot(x='Age',y='EstimatedSalary',data=df,hue="Purchased")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mb=MultinomialNB()
mb.fit(x_train,y_train)
```

```python
mb.score(x_test,y_test)*100,mb.score(x_train,y_train)*100
```

```python
bb=BernoulliNB()
bb.fit(x_train,y_train)
```

```python
bb.score(x_test,y_test)*100,bb.score(x_train,y_train)*100
#
```
